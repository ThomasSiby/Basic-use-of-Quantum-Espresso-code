```python
import matplotlib.pyplot as plt
from matplotlib import rcParamsDefault
import numpy as np

%matplotlib inline

plt.rcParams["figure.dpi"]=150
plt.rcParams["figure.facecolor"]="white"

x, y = np.loadtxt('etot_v_ecut.dat', delimiter=' ', unpack=True)
plt.plot(x, y, "o-", markersize=5, label='Etot vs ecutwfc')
plt.xlabel('ecutwfc (Ry)')
plt.ylabel('Etot (Ry)')
plt.legend(frameon=False)
plt.savefig('etot_v_ecut.jpg', bbox_inches='tight', dpi=600)
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib import rcParamsDefault
import numpy as np

%matplotlib inline

plt.rcParams["figure.dpi"]=150
plt.rcParams["figure.facecolor"]="white"

x, y = np.loadtxt('etot_v_ecut_m1.dat', delimiter=' ', unpack=True)
plt.subplot(1, 2, 1) # row 1, col 2 index 1
plt.plot(x, y, "o-", markersize=5, label='Etot vs ecutwfc')
plt.xlabel('ecutwfc (Ry)')
plt.ylabel('Etot (Ry)')
plt.legend(frameon=False)
plt.savefig('etot_v_ecut_m1.jpg', bbox_inches='tight')
plt.show()

x, y = np.loadtxt('etot_v_ecut_m2.dat', delimiter=' ', unpack=True)
plt.subplot(1, 2, 2) # row 1, col 2 index 1
plt.plot(x, y, "*-", markersize=5, label='Etot vs ecutwfc')
plt.xlabel('ecutwfc (Ry)')
plt.ylabel('Etot (Ry)')
plt.legend(frameon=False)
plt.savefig('etot_v_ecut_m2.jpg', bbox_inches='tight', dpi=600)
plt.tight_layout()
plt.show()

```
